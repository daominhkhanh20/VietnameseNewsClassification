```python
from matplotlib import pyplot as plt 
import pickle 
import numpy as np 
```

```python
with open('/media/daominhkhanh/D:/Data/Project/ViettelProject/Model/Bert/history16.pickle','rb') as file:
    history=pickle.load(file)
```

```python
train_acc=history['train_acc']
train_loss=history['train_loss']
val_acc=history['val_acc']
val_loss=history['val_loss']
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(len(train_acc)),train_acc,label='Train')
plt.plot(range(len(val_acc)),val_acc,label='Val')
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Acc")
plt.savefig('bert_acc.png')
```
